## DATA PRE-PROCESSING

```python
txBitrate=[]
rxBitrate=[]
ulthroughput=[]
dlthroughput=[]

data_preprocessing=pd.DataFrame()
date=np.sort(list(set(list(synchro['Date time']))) )
data_prep=[]
for d in date:
    data_prep.append(pd.to_datetime(d))
    
#from datetime import datetime
#for d in date:
 #   ju.append(datetime.strptime(d, "%Y-%m-%d %H:%M:%S"))
data_preprocessing['Time']=data_prep
for d in date:
    txBitrate.append(np.sum(synchro[synchro['Date time']==d])['txBitrate'])
    rxBitrate.append(np.sum(synchro[synchro['Date time']==d])['rxBitrate'])
    ulthroughput.append(np.sum(synchro[synchro['Date time']==d])['ulthroughput'])
    dlthroughput.append(np.sum(synchro[synchro['Date time']==d])['dlthroughput'])
data_preprocessing['txBitrate']=txBitrate
data_preprocessing['rxBitrate']=rxBitrate
data_preprocessing['ulthroughput']=ulthroughput
data_preprocessing['dlthroughput']=dlthroughput
data_preprocessing
```

```python
import matplotlib as mpl
from matplotlib import cm
import matplotlib.pyplot as plt
from matplotlib.dates import YearLocator, MonthLocator
%matplotlib inline
import seaborn as sns
#import missingno as msno
from tqdm import tqdm
import sklearn.mixture as mix

# Data Normalisation
txBitrate=list(data_preprocessing['txBitrate'])/np.sum(list(data_preprocessing['txBitrate']))
rxBitrate=list(data_preprocessing['rxBitrate'])/np.sum(list(data_preprocessing['rxBitrate']))
ulthroughput=list(data_preprocessing['ulthroughput'])/np.sum(list(data_preprocessing['ulthroughput']))
dlthroughput=list(data_preprocessing['dlthroughput'])/np.sum(list(data_preprocessing['dlthroughput']))

X = np.column_stack([txBitrate,rxBitrate,ulthroughput,dlthroughput])


#Train the model
model = mix.GaussianMixture(n_components=2, 
                            covariance_type="full", 
                            n_init=100, 
                            random_state=7).fit(X)

# Predict the optimal sequence of internal hidden state
hidden_states = model.predict(X)

print("Means and vars of each hidden state")
for i in range(model.n_components):
    print("{0}th hidden state".format(i))
    print("mean = ", model.means_[i])
    print("var = ", np.diag(model.covariances_[i]))
    print()

#sns.set(font_scale=1.25)
style_kwds = {'xtick.major.size': 3, 'ytick.major.size': 3,
              'font.family':u'courier prime code', 'legend.frameon': True}
#sns.set_style('white', style_kwds)
col='rxBitrate'

fig, axs = plt.subplots(model.n_components, sharex=True, sharey=True, figsize=(12,9))
colors = cm.rainbow(np.linspace(0, 1, model.n_components))

for i, (ax, color) in enumerate(zip(axs, colors)):
    # Use fancy indexing to plot data in each state.
    mask = hidden_states == i
    ax.plot_date(data_preprocessing['Time'].values[mask],
                 data_preprocessing[col].values[mask],
                 ".-", c=color)
    ax.set_title("{0}th hidden state".format(i), fontsize=16, fontweight='demi')

    # Format the ticks.
    ax.xaxis.set_major_locator(YearLocator())
    ax.xaxis.set_minor_locator(MonthLocator())
    #sns.despine(offset=10)
    ax.grid(True)
plt.show()
plt.tight_layout()
#fig.savefig('Hidden Markov (Mixture) Model_Regime Subplots.png')
```
